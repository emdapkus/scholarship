```python
#fill NA values in hist df
#group by lat, long, year
df = df.groupby(['long', 'lat', 'year', 'TimePeriod', 'RCP', 'scenario'])
# hist = hist.groupby(['long', 'lat', 'year', 'TimePeriod',  'RCP', 'scenario'])
# near = near.groupby(['long', 'lat', 'year',  'TimePeriod', 'RCP', 'scenario'])

#Fill NA values within each group with non NA values
df = df.apply(lambda x: x.fillna(method = 'ffill').fillna(method = 'bfill'))
# hist = hist.apply(lambda x: x.fillna(method = 'ffill').fillna(method = 'bfill'))
# near = near.apply(lambda x: x.fillna(method = 'ffill').fillna(method = 'bfill'))

#recalculate PPT_annual
df['PPT_Annual'] = df[['PPT_Winter', 'PPT_Spring', 'PPT_Summer', 'PPT_Fall']].sum(axis = 1)
#recalulate T_Annual
df['T_Annual'] = df[['T_Winter', 'T_Spring', 'T_Summer', 'T_Fall']].mean(axis = 1)

#strip white space from strings
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

#round ExtremeShortTermDryStress_Winter_whole to the 5th decimal place
# df = df.drop_duplicates(subset=['ExtremeShortTermDryStress_Winter_whole'])
df.drop(df.round(5).duplicated().loc[lambda x: x].index, inplace=True)
#remove duplicate rows
df = df.drop_duplicates(ignore_index=True)
# hist = hist.drop_duplicates(ignore_index=True)
# near = near.drop_duplicates(ignore_index=True)

#reset index
df = df.reset_index(drop=True)
# hist = hist.reset_index(drop=True)
# near = near.reset_index(drop=True)
```
