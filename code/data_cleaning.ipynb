```python
#import libraries
import pandas as pd
#pandas settings
pd.set_option("display.max_columns", None)
```

```python
#import data
df = pd.read_csv('../data/NABR_ClimExposure', delimiter = ' ', dtype={'RCP': 'category'})
hist = pd.read_csv('../data/NABR_historic.csv', dtype={'RCP': 'category'})
```

```python
#block used to find unique values
col_name = 'scenario'

# Get unique values in the specified column
unique_values = df[col_name].unique()

# Print unique values
print("Unique values in column '{}':".format(col_name))
print(unique_values)
```

```python
#fill NA values in hist df
#group by lat, long, year
df = df.groupby(['long', 'lat', 'year', 'TimePeriod', 'RCP', 'scenario'])

#Fill NA values within each group with non NA values
df = df.apply(lambda x: x.fillna(method = 'ffill').fillna(method = 'bfill'))

#recalculate PPT_annual
df['PPT_Annual'] = df[['PPT_Winter', 'PPT_Spring', 'PPT_Summer', 'PPT_Fall']].sum(axis = 1)
#recalulate T_Annual
df['T_Annual'] = df[['T_Winter', 'T_Spring', 'T_Summer', 'T_Fall']].mean(axis = 1)

#strip white space from strings
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

#round ExtremeShortTermDryStress_Winter_whole to the 5th decimal place
df.drop(df.round(5).duplicated().loc[lambda x: x].index, inplace=True)
#remove duplicate rows
df = df.drop_duplicates(ignore_index=True)
#reset index
df = df.reset_index(drop=True)
```

```python
df.sample(10)
```

```python
#print columns with NANs

# get NANs per col
nan_col = df.isna().sum()
#keep cols with more than 1 nan
nan_cols = nan_col[nan_col > 0]
#print
print(nan_cols)
```

```python
# drop cols with high number of NANs
#find cols with more than 1 NANs ( not planning on using them)
drop_cols = nan_cols > 1
#select cols to drop
drop_cols = drop_cols[drop_cols == True].index
#drop cols
df = df.drop(drop_cols, axis = 1)
```
